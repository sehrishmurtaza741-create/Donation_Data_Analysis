```python
# Remove extra spaces in text columns
text_cols = ['Donation_ID', 'Donor_Gender', 'Donation_Type', 'City', 'Payment_Method', 'Category']
for col in text_cols:
    df[col] = df[col].astype(str).str.strip()

# Convert Donor_Age to numeric, invalid parsing becomes NaN
df['Donor_Age'] = pd.to_numeric(df['Donor_Age'], errors='coerce')

# Convert Donation_Amount to numeric, invalid parsing becomes 0
def clean_amount(x):
    if isinstance(x, str):
        x = x.replace("Rs.", "").replace("zero", "0").replace("NULL", "0").strip()
    try:
        return float(x)
    except:
        return 0

df['Donation_Amount'] = df['Donation_Amount'].apply(clean_amount)

# Remove duplicate rows
df = df.drop_duplicates()

# Fill missing values in other important columns
df['Donor_Gender'] = df['Donor_Gender'].replace(['nan', 'None', ''], 'Unknown')
df['Donation_Type'] = df['Donation_Type'].replace(['nan', 'None', ''], 'Unknown')
df['City'] = df['City'].replace(['nan', 'None', ''], 'Unknown')
df['Payment_Method'] = df['Payment_Method'].replace(['nan', 'None', ''], 'Unknown')
df['Category'] = df['Category'].replace(['nan', 'None', ''], 'Unknown')

# Display cleaned DataFrame
df
```

```python
df.dropna(inplace=True)
df
```

```python
# Total and average donation
total_donation = df['Donation_Amount'].sum()

print("Total Donation Amount:", total_donation)

```

```python
avg_donation = df['Donation_Amount'].mean()
print("Average Donation Amount:", avg_donation)
```

```python
# Donations by city
donation_by_city = df.groupby('City')['Donation_Amount'].sum()
print("\nDonation Amount by City:\n", donation_by_city)
```

```python
# Donations by type
donation_by_type = df.groupby('Donation_Type')['Donation_Amount'].sum()
print("\nDonation Amount by Type:\n", donation_by_type)
```

```python
import matplotlib.pyplot as plt

# Bar chart using Matplotlib (no warnings)
plt.figure(figsize=(8,5))
plt.bar(donation_by_city['City'], donation_by_city['Donation_Amount'], color=plt.cm.viridis(range(len(donation_by_city))))
plt.title("Total Donation Amount by City")
plt.xlabel("City")
plt.ylabel("Donation Amount")
plt.xticks(rotation=45)
plt.show()

```

```python
import matplotlib.pyplot as plt

# Pie chart: Donation Type distribution with better style
donation_type_counts = df['Donation_Type'].value_counts()

colors = ['#ff9999','#66b3ff','#99ff99']  # You can customize
explode = [0.05]*len(donation_type_counts)  # Slightly pull out all slices

plt.figure(figsize=(7,7))
plt.pie(
    donation_type_counts,
    labels=donation_type_counts.index,
    autopct="%1.1f%%",
    startangle=140,
    colors=colors,
    explode=explode,
    shadow=True,
    textprops={'fontsize':12}
)
plt.title("Donation Type Distribution", fontsize=14)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram: Donor Age distribution with better style
plt.figure(figsize=(8,5))
sns.histplot(df['Donor_Age'].dropna(), bins=10, kde=True, color='skyblue', edgecolor='black')
plt.title("Distribution of Donor Ages", fontsize=14)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Number of Donors", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```
